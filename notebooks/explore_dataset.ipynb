```python
# Dataset desde Spaces: procurements_dataset.json → pandas
# Alternativa (archivo local): pd.read_json("/Users/kukiamarilla/Downloads/procurements_dataset.json")

import io
import sys
from pathlib import Path

import pandas as pd
from dotenv import load_dotenv

# Raíz del repo (asumiendo que el kernel corre con CWD = notebooks/)
REPO_ROOT = Path.cwd().resolve()
if REPO_ROOT.name == "notebooks":
    REPO_ROOT = REPO_ROOT.parent

sys.path.insert(0, str(REPO_ROOT / "scripts" / "etl"))
load_dotenv(REPO_ROOT / ".env")

import spaces_io

client = spaces_io.s3_client()
bucket = spaces_io.bucket_name()
dataset_key = spaces_io.object_key(
    spaces_io.dataset_prefix(),
    "procurements_dataset.json",
)

body = spaces_io.get_object_bytes(client, bucket, dataset_key)
df = pd.read_json(io.BytesIO(body))

print(f"{len(df)} filas × {len(df.columns)} columnas  ←  s3://{bucket}/{dataset_key}")
df.head()
```
